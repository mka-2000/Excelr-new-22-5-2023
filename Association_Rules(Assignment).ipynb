```python
# Q1) Books.csv 

# We are given with the dataset of Books.
# We are required to find the association of the books.

# First we define the Antescedent and Consequent products.

# In this dataset:

#  Antescedent Product  :  The Book which we actually wanted to buy.

# Consequent Product    :  The extra books that are bought other than the main book.

!pip install mlxtend

import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
book = pd.read_csv('book.csv')
print(book)
book.head(10)
```

```python
# Apriori Algorithm
# To select the min_support value , we check different combination of values.
```

```python
freq_items = apriori(book,min_support=0.5,use_colnames=True)
freq_items
```

A min_support of 0.5 is said to be a good value.but in this dataset, we do not have any values with support of 0.5 and above.

```python
freq_items = apriori(book,min_support=0.4,use_colnames=True)
freq_items
```

It has only 2 datas in it.So,going for 0.3

```python
freq_items = apriori(book,min_support=0.3,use_colnames=True)
freq_items
```

Again only 2 values , we are getting.So, go for 0.2

```python
freq_items = apriori(book,min_support=0.2,use_colnames=True)
freq_items
```

OK!   0.2 seems to be a good support value.But let us check for 0.1

```python
freq_items = apriori(book,min_support=0.1,use_colnames=True)
freq_items
```

As we can see that 0.1 value of minimum support gives maximum values , we take 0.1 as the confidence value and do our calculations further.

```python
# After doing trail and Error , we take min support to be 0.1
```

```python
freq_items = apriori(book,min_support=0.1,use_colnames=True)
freq_items
```

```python
rules = association_rules(freq_items,metric="lift",min_threshold=0.7)
rules
```

Leaverage value ranges between -1 to 1.

A leaverage value of 0 indicates independence.

```python
rules.sort_values('lift',ascending=False).head(15)
```

```python
rules.sort_values('lift',ascending=False)[0:20]
```

```python
rules[rules.lift>1]
```

From the above table, 

--> we see that the " Lift " values are greater than 1.

-> This means that the dependency of the Consequent to the Antescedent is very High.

--> Therefore, they can say that the books that are bought by each customers share a strong association between themselves.
